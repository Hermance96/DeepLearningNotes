```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras import models
from keras import layers
from keras.models import load_model
from keras.preprocessing import image
```

# 可视化中间激活

```python
model = load_model('cvd_small2.h5')
model.summary()
```

```python
img_path = '/Users/liuxuanchi/Downloads/cats_and_dogs_small/test/cats/cat.1700.jpg'

# target_size: img_height, img_width
img = image.load_img(img_path, target_size=(150, 150))
# 将PIL图像实例转换为Numpy数组
img_tensor = image.img_to_array(img)
# 扩展维度，加入一个新的轴axis
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

print(img_tensor.shape)
```

```python
plt.imshow(img_tensor[0])
plt.show()
```

```python
layer_outputs = [layer.output for layer in model.layers[:8]]
act_model = models.Model(inputs=model.input, outputs=layer_outputs)
```

```python
acts = act_model.predict(img_tensor)
first_layer_act = acts[0]
print(first_layer_act.shape)
```

```python
# cmap选择颜色图
plt.matshow(first_layer_act[0, :, :, 4], cmap='viridis')
```

```python
plt.matshow(first_layer_act[0, :, :, 7])
```

```python
layer_names = []
for layer in model.layers[:8]:
    layer_names.append(layer.name)

images_per_row = 16

for layer_name, layer_act in zip(layer_names, acts):
    # layer_act in form of (1, size, size, features)
    n_ftrs = layer_act.shape[-1]
    size = layer_act.shape[1]
    n_cols = n_ftrs // images_per_row
    display_grid = np.zeros((size*n_cols, size*images_per_row))
    
    for col in range(n_cols):
        for row in range(images_per_row):
            channel_img = layer_act[0, :, :, col*images_per_row+row]
            channel_img -= channel_img.mean()
            channel_img /= channel_img.std()
            channel_img *= 64
            channel_img += 128
            # clip(array, min, max)，将数组中元素限制在[min,max]范围内，超出范围的一律看作上/下界
            channel_img = np.clip(channel_img, 0, 255).astype('uint8')
            display_grid[col*size : (col+1)*size, row*size : (row+1)*size] = channel_img
    
    scale = 1. / size
    plt.figure(figsize=(scale*display_grid.shape[1],
                       scale*display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
```

# 过滤器可视化 

```python
from keras.applications import VGG16
from keras import backend as K

model = VGG16(weights='imagenet',
             include_top=False)

# 'block3_conv1'层
layer_name = 'block3_conv1'
filter_idx = 0

layer_output = model.get_layer(layer_name).output
# 第0个过滤器激活的损失
loss = K.mean(layer_output[:, :, :, filter_idx])

# 损失相对输入的梯度
grads = K.gradients(loss, model.input)[0]

# 梯度标准化
grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5)
```

```python
iterate = K.function([model.input], [loss, grads])

loss_val, grads_val = iterate([np.zeros((1, 150, 150, 3))])
```

```python
input_img = np.random.random((1, 150, 150, 3)) * 20 + 128.
step = 1.
for i in range(40):
    loss_val, grads_val = iterate([input_img])
    input_img += grads_val*step
```

```python
# tensor -> image
def deprocess_image(x):
    x -= x.mean()
    x /= (x.std() + 1e-5)
    x *= 0.1
    
    x += 0.5
    x = np.clip(x, 0, 1)
    
    x *= 255
    x = np.clip(x, 0, 255).astype('uint8')
    
    return x
```

```python
def generate_pattern(layer_name, filter_idx, size=150):
    layer_output = model.get_layer(layer_name).output
    
    loss = K.mean(layer_output[:, :, :, filter_idx])
    grads = K.gradients(loss, model.input)[0]
    grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5)
    
    iterate = K.function([model.input], [loss, grads])
    
    input_img = np.random.random((1, size, size, 3)) * 20 + 128.
    
    step=1.
    for i in range(40):
        loss_val, grads_val = iterate([input_img])
        input_img += grads_val * step
    
    img = input_img[0]
    return deprocess_image(img)
```

```python
plt.imshow(generate_pattern('block3_conv1', 0))
```

```python
layer_name = 'block1_conv1' 
size = 64
margin = 5

results = np.zeros((8 * size + 7 * margin, 8 * size + 7 * margin, 3))

for i in range(8):
    for j in range(8):
        filter_img = generate_pattern(layer_name, i + (j * 8), size=size)
        horizontal_start = i * size + i * margin 
        horizontal_end = horizontal_start + size 
        vertical_start = j * size + j * margin 
        vertical_end = vertical_start + size 
        results[horizontal_start: horizontal_end,
                vertical_start: vertical_end, :] = filter_img

plt.figure(figsize=(20, 20))
plt.imshow(results)
```
